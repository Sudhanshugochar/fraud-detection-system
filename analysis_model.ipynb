```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
df.head()
```

```python
df_model = df.drop(["nameOrig", "nameDest", "isFlaggedFraud"], axis=1)
```

```python
categorical = ["type"]
numeric = ["amount", "oldbalanceOrg", "newbalanceOrig", "oldbalanceDest", "newbalanceDest"]
```

```python
# 1. Define Target (y) and Features (X)
y = df_model["isFraud"]
X = df_model.drop("isFraud", axis=1)

# 2. Split the Data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y)

# 3. Setup the Preprocessing Pipeline
preprocessor = ColumnTransformer(
    # TYPO CORRECTION: 'transformers' instead of 'transofrnmers'
    transformers=[ 
        ("num", StandardScaler(), numeric),
        ("cat", OneHotEncoder(handle_unknown='ignore', sparse_output=False), categorical) # Added robust parameters
    ],
    remainder="drop"
)
```

```python
pipeline = Pipeline(
    [
        ("prep", preprocessor),
        ("clf", LogisticRegression(class_weight="balanced", max_iter=1000))
    ]
)
```

```python
pipeline.fit(X_train, y_train)
```

```python
y_pred = pipeline.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
confusion_matrix(y_test, y_pred)
```

```python
pipeline.score(X_test, y_test) * 100
```
